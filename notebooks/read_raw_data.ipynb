```python
%%time
import pandas as pd
import json
from tqdm import tqdm
import pyarrow as pa
import pyarrow.parquet as pq
```

```python
# df = pd.read_json("../data/raw/arxiv-metadata-oai-snapshot.json")

# Initialize an empty list to store the data from the JSON file
arxiv_data = []

# Open the JSON file using the 'with' statement, ensuring it's closed automatically
with open("/Users/luis.morales/Desktop/arxiv-paper-recommender/data/raw/arxiv-metadata-oai-snapshot.json", 'r') as file:
    # Iterate over each line in the file
    for line in tqdm(file):
        # Load the JSON data from each line and append it to the arxiv_data list
        arxiv_data.append(json.loads(line))

# Create a DataFrame from the arxiv_data list using the pd.DataFrame.from_records() method
df = pd.DataFrame.from_records(arxiv_data)
```
